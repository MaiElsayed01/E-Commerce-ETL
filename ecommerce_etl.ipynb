Load CSV Files

```python
import pandas as pd

user_df = pd.read_csv('recource/user_dataset.csv')
seller_df = pd.read_csv('recource/seller_dataset.csv')
product_df = pd.read_csv('recource/products_dataset.csv')
order_df = pd.read_csv('recource/order_dataset.csv')
order_item_df = pd.read_csv('recource/order_item_dataset.csv')
payment_df = pd.read_csv('recource/payment_dataset.csv')
feedback_df = pd.read_csv('recource/feedback_dataset.csv')
```

Transormations

```python
product_df['product_photos_qty'] = product_df['product_photos_qty'].astype('Int16')
product_df['product_description_lenght'] = product_df['product_description_lenght'].astype('Int16')
product_df['product_name_lenght'] = product_df['product_name_lenght'].astype('Int16')
product_df.info()
```

```python
user_df['customer_zip_code'] = user_df['customer_zip_code'].astype(str)
seller_df['seller_zip_code'] = seller_df['seller_zip_code'].astype(str)
user_df.info()
seller_df.info()
```

```python
feedback_df['feedback_answer_date'] = pd.to_datetime(feedback_df['feedback_answer_date'])
feedback_df['feedback_form_sent_date'] = pd.to_datetime(feedback_df['feedback_form_sent_date'])
feedback_df.info()
```

```python
order_df = order_df.drop(columns=['order_approved_date'])
order_df.head()
```

```python
order_df['order_date'] = pd.to_datetime(order_df['order_date'])
order_df['order_hour'] = order_df['order_date'].dt.hour
order_df['order_date'] = order_df['order_date'].dt.date

order_df['delivered_date'] = pd.to_datetime(order_df['delivered_date'])
order_df['delivered_date'] = order_df['delivered_date'].dt.date

order_df['estimated_time_delivery'] = pd.to_datetime(order_df['estimated_time_delivery']) 
order_df['estimated_time_delivery'] = order_df['estimated_time_delivery'].dt.date

order_df['pickup_date'] = pd.to_datetime(order_df['pickup_date'])
order_df['pickup_date'] = order_df['pickup_date'].dt.date

order_df.info()
```

```python
order_item_df = order_item_df.drop(columns=['pickup_limit_date'])
order_item_df = order_item_df.drop(columns=['order_item_id'])

order_item_df.head()
```

```python
order_item_df['price'] = order_item_df['price'].round(2)
order_item_df['shipping_cost'] = order_item_df['shipping_cost'].round(2)
order_item_df.head()
```

```python
payment_df['payment_value'] = payment_df['payment_value'].round(2)
payment_df.head()
```

Merge order_item with order

```python
order_df = pd.merge(order_df, order_item_df, on='order_id', how='inner')
order_df.head(10)
```

Add Surrogate Key

```python
user_df = user_df.reset_index()
product_df = product_df.reset_index()
seller_df = seller_df.reset_index()
payment_df = payment_df.reset_index()
```

Replace product id (BK) with product key (SK) in Order DF

```python
order_df = pd.merge(order_df, product_df[['product_id', 'index']], on='product_id', how='inner')
order_df = order_df.rename(columns={'index': 'product_key'})
order_df = order_df.drop(columns=['product_id'])
order_df.head()
```

Replace user id (BK) with user key (SK) in Order

```python
order_df = pd.merge(order_df, user_df[['user_name', 'index']], on = 'user_name', how='inner')
order_df = order_df.rename(columns={'index': 'user_key'})
order_df = order_df.drop(columns=['user_name'])
order_df.head()
```

Replace seller id (BK) with seller key (SK)

```python
order_df = pd.merge(order_df, seller_df[['seller_id', 'index']], on='seller_id', how='inner')
order_df = order_df.rename(columns={'index': 'seller_key'})
order_df = order_df.drop(columns=['seller_id'])
order_df.head()
```

Replace payment id (BK) with payment key (SK)

```python
order_df = pd.merge(order_df, payment_df[['order_id', 'index']], on='order_id', how='inner')
order_df = order_df.rename(columns={'index': 'payment_key'})
order_df.head()
```

Create Date and Time DF

```python
start_date = order_df['order_date'].min()
end_date = order_df['order_date'].max()

date_df = pd.DataFrame(pd.date_range(start_date, end_date), columns=['date'])
date_df['year'] = date_df['date'].dt.year
date_df['quarter'] = date_df['date'].dt.quarter
date_df['month'] = date_df['date'].dt.month
date_df['day'] = date_df['date'].dt.day
date_df['date'] = date_df['date'].dt.date
date_df = date_df.reset_index()
date_df = date_df.rename(columns={'index': 'date_key'})

time_df = pd.DataFrame(range(24), columns=['hour'])
time_df = time_df.reset_index()
time_df = time_df.rename(columns={'index': 'time_key'})
time_df.head(20)
```

Replace Dates in Order with Date Key

```python
order_df = pd.merge(order_df, date_df[['date_key','date']], right_on='date', left_on='order_date', how='inner')

order_df = order_df.drop(columns=['order_date'])
order_df = order_df.drop(columns=['date'])

order_df = order_df.rename(columns={'date_key': 'order_date_key'})

order_df.head()
```

```python
order_df = pd.merge(order_df, time_df[['time_key','hour']], right_on='hour', left_on='order_hour', how='inner')

order_df = order_df.drop(columns=['order_hour'])
order_df = order_df.drop(columns=['hour'])

order_df = order_df.rename(columns={'time_key': 'order_time_key'})

time_df.head()
```

```python
order_df = pd.merge(order_df, date_df[['date_key','date']], right_on='date', left_on='pickup_date', how='inner')

order_df = order_df.drop(columns=['pickup_date'])
order_df = order_df.drop(columns=['date'])

order_df = order_df.rename(columns={'date_key': 'pickup_date_key'})

order_df.head()
```

```python
order_df = pd.merge(order_df, date_df[['date_key','date']], right_on='date', left_on='delivered_date', how='inner')

order_df = order_df.drop(columns=['delivered_date'])
order_df = order_df.drop(columns=['date'])

order_df = order_df.rename(columns={'date_key': 'delivered_date_key'})

order_df.head()
```

```python
order_df = pd.merge(order_df, date_df[['date_key','date']], right_on='date', left_on='estimated_time_delivery', how='inner')

order_df = order_df.drop(columns=['estimated_time_delivery'])
order_df = order_df.drop(columns=['date'])

order_df = order_df.rename(columns={'date_key': 'estimated_time_delivery_key'})

order_df.head()
```

add SK for order

```python
order_df = order_df.reset_index()
order_df.head()
```

Remove Order ID from Payment

```python
payment_df = payment_df.drop(columns=['order_id'])
payment_df.head()
```

Replace order_id with order index in feedback

```python
feedback_df = pd.merge(feedback_df, order_df[['order_id', 'index']], on='order_id', how='inner')

feedback_df = feedback_df.rename(columns={'index': 'order_key'})
feedback_df = feedback_df.drop(columns=['order_id'])

feedback_df.head()
```

Load Data into DWH

```python
from sqlalchemy import create_engine
from config import server, database, driver 

connection_string = f"mssql+pyodbc://@{server}/{database}?driver={driver}"

engine = create_engine(connection_string)
```

```python
user_df = user_df.rename(columns={'index': 'user_key', 'user_name': 'user_id', 'customer_zip_code': 'user_zip_code', 'customer_city': 'user_city', 'customer_state': 'user_state'})
user_df.to_sql('user_dim', con=engine, if_exists='append', index=False)
```

```python
seller_df = seller_df.rename(columns={'index': 'seller_key'})
seller_df.to_sql('seller_dim', con=engine, if_exists='append', index=False)
```

```python
product_df = product_df.rename(columns={'index': 'product_key'})
product_df = product_df.rename(columns={'product_name_lenght': 'product_name_length', 'product_description_lenght': 'product_description_length'})
product_df.to_sql('product_dim', con=engine, if_exists='append', index=False)
```

```python
payment_df = payment_df.drop(columns=['payment_sequential'])
payment_df = payment_df.rename(columns={'index': 'payment_key'})
payment_df.to_sql('payment_dim', con=engine, if_exists='append', index=False)
```

```python
date_df.to_sql('date_dim', con=engine, if_exists='append', index=False)
```

```python
time_df.to_sql('time_dim', con=engine, if_exists='append', index=False)
```

```python
order_df = order_df.rename(columns={'index': 'order_key'})
order_df.to_sql('order_fact', con=engine, if_exists='append', index=False)
```

```python
feedback_df = feedback_df.drop_duplicates(subset='feedback_id')
feedback_df.to_sql('feedback_dim', con=engine, if_exists='append', index=False)
```
